```python
import pandas as pd
df=pd.read_csv("netflix_titles_1.csv")
df
```

```python
print("Information of DataSet: Summary")
df.info()
```

```python
print("The shape of Dataset: ", df.shape)
print("The size of Dataset: ", df.size)
print("Datatypes: \n", df.dtypes)
```

```python
pd.set_option('display.max_columns', None)
print("Describe Dataset")
df.describe(include='all')
```

```python
print("The number of NaN values in each column: ")
df.isna().sum()
```

```python
print("The number of null values in each column: ")
df.isnull().sum()
```

```python
df.groupby('type')['rating'].count()
```

```python
c=['type', 'country', 'rating', 'duration']
for column in c:
    print(f"Column: {column}")
    print(df[column].value_counts())
    print("\n")
```

```python
values = ["1 Season", "2 Seasons", "3 Seasons"]
df2 = df[df['duration'].isin(values)]
df2
```

```python
m_df = df[df['type'] == 'Movie']
ts_df = df[df['type'] == 'TV Show']
movies_country = m_df.groupby('country').size().reset_index(name='movies_count')
tv_shows_country = ts_df.groupby('country').size().reset_index(name='tv_shows_count')
print("Movies Count by Country:")
print(movies_country)
print("\nTV Shows Count by Country:")
print(tv_shows_country)
```
