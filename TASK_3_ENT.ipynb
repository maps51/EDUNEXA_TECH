TASK 3: Data Cleaning and Preprocessing.
Select a dataset with missing values and outliers. Apply techniques to clean and preprocess the data using Pandas. Impute missing values and handle outliers appropriately.

```python
import pandas as pd
df=pd.read_csv("netflix_titles_1.csv")
df
```

```python
print("NaN values:")
df.isna()
```

```python
print("The number of NaN values per column: ")
df.isna().sum()
```

```python
print("Total NaN values: ", df.isna().sum().sum())
```

```python
df2=df.dropna()
print("Dataset after dropping NaN values: ",df2.shape)
df2
```

```python
df3=df.fillna(0)
print("Filling missing data with zero: ", df3.shape)
df3
```

```python
df4=df
df4['director'].fillna("Not Mentioned", inplace=True)
df4['cast'].fillna("Not Mentioned", inplace=True)
df4['country'].fillna("Not Mentioned", inplace=True)
print("The NaN values in this dataset: ", df4.isna().sum().sum())
df4
```

```python
df4['date_added'] = pd.to_datetime(df4['date_added'].astype(str).str.replace(',', ''), errors='coerce')
df4['date_added'] = df4['date_added'].fillna('Not Available')
df4
```

```python
df4.dropna(how='all')
```

```python
df4.isna().sum()
```

```python
df4['rating'].fillna("No rating", inplace=True)
df4['duration'].fillna("Not specified", inplace=True)
print("The number of NaN values: ",df4.isna().sum().sum())
df4
```

For outliers, the released_year column is being considered which is a numerical value. So the conversion of this column to string is done.

```python
df4['release_year']=df4['release_year'].astype(str)
df4
```
